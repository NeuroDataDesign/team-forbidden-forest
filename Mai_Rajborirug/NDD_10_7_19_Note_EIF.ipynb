#### Background: Isolation Forest (IF)
Isolation Forest (IF) algorithm: the anomolies(outliers) usually are rare and further from norms. We judge anomoly score from from: $$s(x,n)=2^{-\frac{\mathbb{E}(h(X))}{c(n)}} \text{  when,}$$ 
- $\mathbb{E}(h(X))$ is the length of edge from the root node to the terminal node
- $c(n)$ is the average length
- $x$ is the data point
- $n$ is sample population
- $0<s(x,n)<1$: the closer $s(x,n)$ to $1$ the more possibility the data is to be an outlier.
